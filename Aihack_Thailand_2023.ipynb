```python

train_data.fillna(train_data.median(), inplace=True)
public_dataset.fillna(public_dataset.median(), inplace=True)

# ---------------------------------------------- Feature Engineering -----------------------------------------------------

train_data['living_period_overall'] = train_data['living_period_month'] + train_data['living_period_year']*12
public_dataset['living_period_overall'] = public_dataset['living_period_month'] + public_dataset['living_period_year']*12

train_data['c_number_of_working_overall'] = train_data['c_number_of_working_month'] + train_data['c_number_of_working_year']*12
public_dataset['c_number_of_working_overall'] = public_dataset['c_number_of_working_month'] + public_dataset['c_number_of_working_year']*12

# ----------------------------------------------- Selecting columns ------------------------------------------------------ , 'c_number_of_working_overall'

cols = ['FICO_Score', 'c_occupation','c_position', "c_number_of_working_overall", 'living_period_overall'
          , 'r_expected_credit_limit', 'r_propose', 'c_employment_status', 'r_spouse_income', 'r_allloan_case'
          , 'c_postal_code', 'number_of_children', 'number_of_resident', 'r_allloan_amount'
          , 'c_date_of_salary_payment', 'type_of_residence', 'media', 'r_additional_income', 'r_generalcode1', 'r_generalcode2'
          , 'r_generalcode3', 'c_salary_payment_methods']

df_x = train_data[cols]
df_y = train_data['default_12month']

train_x, test_x, train_y, test_y = train_test_split(df_x, df_y, test_size=0.1)
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from numpy import arange

lda = LinearDiscriminantAnalysis(shrinkage = 0.0, solver = 'lsqr')
lda.fit(train_x, train_y)

'''
## Using GridSearch

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

grid = dict()
grid['solver'] = ['svd', 'lsqr', 'eigen']
grid['shrinkage'] = arange(0, 1, 0.01)

search = GridSearchCV(lda, grid, scoring='accuracy', cv=cv, n_jobs=-1)

results = search.fit(train_x, train_y)

print('Mean Accuracy: %.3f' % results.best_score_)
print('Config: %s' % results.best_params_)

'''
```
